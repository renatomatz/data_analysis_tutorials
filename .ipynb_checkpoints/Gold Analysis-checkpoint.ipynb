## Function Definitions

In this analysis I will use custom functions defined here for the sake of cleanliness later on in the notebook. 

These will do the following:

* Implementing a model.

* Condense some of the processing needed to fit and train a model.

* Process model outputs. 

* Graph model outputs.

These might not have much context as they are, so skipping this part is an option and first understanding how they are used.

```python
def risk_metrics(data, lam, ignore_first=True): 
    """Get period volatitlity for a a dataframe or series."""
    
    data = data.copy()**2
    last_ret = data.iloc[int(ignore_first)].copy()

    for i in range(1 + int(ignore_first), data.shape[0]):
        curr = data.iloc[i].copy()
        data.iloc[i] = lam*(data.iloc[i-1]) + (1-lam)*(last_ret)
        last_ret = curr

    return data
```

```python
def get_var(arch_model, 
            quantiles=[0.025, 0.05],
            fit_params=None):
    """Get the value at risk data for a ready ARCH model"""
    
    if fit_params is not None:
        if not isinstance(fit_params, dict):
            raise TypeError("fit_params arguments must be a dict")
    else:
        fit_params = dict()
    
    fit_res = arch_model.fit(**fit_params)

    # prepare returns data
    returns = arch_model._y_original
    while type(returns.columns).__name__ == "MultiIndex":
        returns = returns.droplevel(1, axis=1)
    returns.columns = ["returns"]

    # start is the earliest index fitted by the algorithm
    start = returns.index[max(0, fit_res._fit_indices[0]-1)]

    if returns is None:
        raise ValueError("Invalid ARCH model: does not have returns")
    
    forecasts = fit_res.forecast(start=start)

    cond_mean = forecasts.mean[start:]
    cond_var = forecasts.variance[start:]

    dist = type(am.distribution).__name__

    # set ppf parameters according to data distribution
    if dist == "Normal":
        params = None
    elif dist == "StudentsT":
        params = fit_res.params[-1]
    elif dist == "SkewStudent":
        params = fit_res.params[-2:]
    else:
        raise TypeError(f"model has unsuported distribution {dist}")

    q = am.distribution.ppf(quantiles, params)

    value_at_risk = -cond_mean.values \
        - np.sqrt(cond_var).values \
        * q[None, :]

    value_at_risk = pd.DataFrame(
        value_at_risk,
        columns=quantiles,
        index=cond_var.index
    )

    max_exedence = []
    for idx in value_at_risk.index:
        for quantile in sorted(quantiles):
            if float(-returns.loc[idx]) > value_at_risk.loc[idx, quantile]:
                max_exedence.append(quantile)
                break
        else:
            # this only executes if the whole for loop completes
            max_exedence.append(1)
            
    max_exedence = pd.DataFrame(
        max_exedence,
        columns=["max_exedence"],
        index=cond_var.index
    )
    
    return pd.concat(
        [returns, value_at_risk, max_exedence],
        join="inner",
        axis=1
    ).dropna()

def plot_var(var, ax=None):
    """Visualize value at risk data and outliers."""
             
    # plot returns
    for exedence, data in var.groupby("max_exedence"):
        ax.scatter(data.index, -data.returns.values,
                   alpha=0.5,
                   label=exedence)

    var = var.drop(["returns", "max_exedence"], axis=1)
    
    # plot value at risk lines
    for col, data in var.iteritems():
        ax.plot(data.index, data.values, 
                linewidth=0.5,
                alpha=0.6,
                label=col)

    ax.legend()
```

```python
def ts_to_labels(_ts, _n, col=None):
    """Create a set of labels from a time series. This is used to fit time series prediction models."""
    _ts = _ts if isinstance(_ts, pd.DataFrame) \
        else pd.DataFrame(_ts, columns=["x"])

    _x, _y = list(), list()
    _ts.rolling(_n+1).apply(append_window,
                            args=(_x, _y, _n),
                            kwargs={"col": col})

    return np.array(_x), np.array(_y)


def append_window(_w, _x, _y, _n, col=None):
    """A helper method to append a time window onto a list."""
    _x.append(np.array(_w.iloc[:_n]))
    _y.append(np.array(_w.iloc[_n]) if col is None
              else np.array(_w.iloc[_n][col]))

    return 1
    
def fit_ts_model(model, 
                 _ts,  
                 _n,
                 *args,
                 fit_attr="fit", 
                 shuffle=True,
                 **kwargs):
    """Fit custom time series prediction model based on provided data."""

    if (len(_ts.shape) == 2 and _ts.shape[1] != 1) \
    or (len(_ts.shape) > 2):
        raise ValueError(f"input time series must be a 1D array, not \
            {len(_ts.shape)}D")

    _x, _y = ts_to_labels(_ts, _n)
    _y = _y.reshape(-1, 1)
    
    if shuffle:
        concat = np.concatenate((_x, _y), axis=1)
        np.random.shuffle(concat)
        _x, _y = np.split(concat, [_n], axis=1)

    model.__getattribute__(fit_attr)(
        _x,
        _y.reshape(-1, 1),
        *args,
        **kwargs
    )

def predict_ts_model(model, 
                     _ts, 
                     _n,
                     *args, 
                     start=None, 
                     horizon=1,
                     pred_attr="predict",
                     **kwargs):
    """Predict the next steps in a time series given data and a fitted time series prediction model"""

    if (len(_ts.shape) == 2 and _ts.shape[1] != 1) \
    or (len(_ts.shape) > 2):
        raise ValueError(f"""input time series must be a 1D array, not
            {len(_ts.shape)}D""")
        
    if isinstance(_ts, pd.DataFrame):
        _ts = _ts.values.reshape(-1)

    if len(_ts) < _n:
        ValueError(f"input musut have at least {_n} items.")

    ret_x, ret_pred = [], []

    curr_x, curr_pred = np.empty(0), None
    if start is None:
        curr_x = _ts[-_n:]
    else:
        if len(_ts[start:]) < _n:
            ValueError(f"specify a start with more than {_n} items \
                ahead of it.")
        curr_x = _ts[start:start+_n]

    for _ in range(horizon):
        curr_pred = model.__getattribute__(pred_attr)(
            np.array(curr_x.reshape(-1, _n)),
            *args,
            **kwargs
        )
            
        ret_x.append(curr_x.copy())
        ret_pred.append(curr_pred)
        curr_x[:_n-1], curr_x[_n-1] = curr_x[1:], curr_pred

    return np.array(ret_x), np.array(ret_pred)
```

## Importing data

Here we import data from standardized files. Keep in mind these were downloaded from the same source and are named in a convenient way to allow for this specific importing procedure. 

Starting to plan out your data is a great way of stepping up your analyses' size efficiency. This naming schedule, for example, allows us to very intuitively change the years list to import only a subset of the gold price years.

It's important to mention the importance of knowing the data before importing it this way. If you are dealing with a source for the first time, make sure you inspect and understand it well before taking the risk of creating errors during a cleaner import.

```python
# The data is composed of Gold Futures data acquired by investing.com
# https://www.investing.com/commodities/gold

gold_all =  pd.concat([
    pd.read_csv("gold"+year+".csv", 
                usecols= ["Date", "Price"],
                dtype= {"Date": "str", "Price": "str"})
    for year in ["1980", "1990", "2000", "2010", "2020"]
])

gold_all["Date"] = gold_all["Date"].apply(pd.Timestamp)
gold_all["Price"] = ( 
    gold_all["Price"]
    .str.replace(",", "")
    .astype(float)
)

df = gold_all.set_index("Date")
df
```

## Time series graphical overview

Once you apply some summary statistics to your data to get a good numerical sense of it, visualizations are your next go-to tool to spot patterns or irregularities.

Keep in mind that while it is very tempting at any one step in the analysis to look for patters and insights right away, keeping a sharp lookout for flaws in the data or your processing is key for making your insights _actually useful_. This is harder than it seems, as what can look like an incredible pattern can be an error, and seemingly corrupt data can actually be something worth exploring.

```python
fig = plt.figure(figsize=(12, 8))
gs = gridspec.GridSpec(2, 2)

ax1 = fig.add_subplot(gs[0, 0])
df.pct_change().rename({"Price": "Percentage Change"}, axis=1).plot(ax=ax1)
ax1.set_title("Daily Percentage Change")

ax2 = fig.add_subplot(gs[0, 1])
df.iloc[100:].pct_change().pipe(risk_metrics, 0.94).rename({"Price": "Volatility"}, axis=1).plot(ax=ax2)
"""
My decision to remove the first 100 items reflects my comment on keeping a lookout for errors.

Notice that keeping these first few data points causes what seems to be a peak of volatility right at 
the beginning of the dataset. 

When observing this, one should ask whether this is an indication of some sort of error on the part of the
analyst (maybe the risk_metrics() function is off) or just some quirk of gold prices at the beginning of the 80s. 
This is where other visualizations, deeper analysis of your data, your algorithms or even
research into the history of the dataset comes in.

I left these few data points out by default to make other volatility spikes clearer in the graph, but 
ask yourself: "does this decision hide one important insight at the cost of another?"

Hint: Look at the other two graphs at the same period.
"""
ax2.set_title("Daily Volatility (lambda = 0.94)")

ax3 = fig.add_subplot(gs[1, :])
df.plot(ax=ax3)
ax3.set_title("Daily Prices")
```

## Ordinary Least Squares model

Dealing with large datasets is fairly dangerous as small errors can corrupt your models, and these will only cause the program itself to fail only if you're lucky; common sense and appreciation of modeling techniques is key to spotting these errors and fixing them.

Goint into the appreciation of the example models I use here would be beyond the objectives of this tutorial, so just realize that statistical diagnostic tools are just important as actual results to ensure your analysis is solid. Know the software you are using and what the available tools are, and most importantly, know how to trace back possible errors based on these diagnostics and ultimately improve your model and thus your results. 

```python
x = np.arange(len(df))
y = df.values

lm_res = sm.OLS(y, sm.add_constant(x)).fit()
```

```python
lm_res.summary()
```

```python
prstd, upper, lower = wls_prediction_std(lm_res)
lm_fig, lm_ax = plt.subplots(1, figsize=(12, 8))

lm_ax.scatter(x, y, s=3)
lm_ax.plot(x, lm_res.fittedvalues)
lm_ax.plot(x, upper, "r")
lm_ax.plot(x, lower, "r")

lm_fig.show()
```

```python
am = aum.ConstantMean(df.pct_change().apply(lambda x: x*100).dropna())
am.volatility = auv.GARCH()
am.distribution = aud.Normal()

arch_res = am.fit()
```

```python
var = get_var(am, quantiles=[0.025, 0.1])
var
```

## Time Series Prediction with Neural Networks

Building a serious neural network using this data would require significant data augmentation and manipulation, which would go beyond the point of this tutorial. Just the simple functions to split training data and make predictions took up considerable space, so this should be enough to show that how good is the data is just as important as how good is the model. 

Notice how easily a model can overfit an un-augmented dataset, even if there are not many data points. Even the simple Multi-Layer Perceptron model used bellow clearly can't predict more than a few sane gold returns ahead.

This is not to say neural networks can't be used for time series prediction, but instead, its a stark reminder than when you start messing with these, the model itself starts becoming a lesser worry.

```python
pred_context = 7

km = Sequential()

km.add(Dense(50, activation="linear", input_shape=(pred_context,)))
km.add(Dropout(0.5))
km.add(Dense(1))

km.compile(loss="mse",
           optimizer="adam",
           metrics=["mse"])
```

```python
fit_ts_model(
    km, df.pct_change().dropna(), pred_context, shuffle=True,
    batch_size=10, epochs=8
)
```

```python
_, res = predict_ts_model(
    km, df.pct_change().dropna(), pred_context,
    horizon=10
)
res
```
